*Précision* : vous construirez un vecteur $\tilde X$ de $N$ points équi-répartis entre $-5$ et $5$, le vecteur $\tilde Y = f(\tilde X)$ avec $f:x\mapsto e^{-x^2}$. Puis vous définirez $X$ et $Y$ des vecteurs de taille $N$ tels que 
$$
X_k = \tilde X_k + \epsilon \delta_k^X, 
\quad
Y_k = \tilde Y_k + \epsilon \delta_k^Y,
\quad
0\leq k\leq N-1,
$$
où $\delta_k^X$ et $\delta_k^Y$ sont des tirages aléatoires (avec une loi uniforme par exemple) de valeurs entre $-1$ et $1$. Le choix de $\epsilon$ est laissé à votre convenance et vous pourrez tester différentes valeurs.

## Les polynômes interpolateurs de Lagrange

La technique d'interpolation par des polynômes interpolateurs de Lagrange repose sur le théorème suivant.

> **Théorème (Polynôme interpolateur de Lagrange)**<br>
> Soient $(X_1,\ldots,X_N)$ et $(Y_1,\ldots Y_N)$ tels que $X_i\neq X_j$ si $i\neq j$.<br>
  Il existe un unique polynôme $P\in\mathbb{R}_{N-1}[X]$ tel que $P(X_i)=Y_i$ pour tout $1\leq i\leq N$.
  
De nombreuses méthodes existent pour déterminer le polynôme interpolateur de Lagrange et surtout l'évaluer en un grand nombre de points (pour le tracer par exemple). Nous choisissons ici la plus simple et la moins efficace (car elle est utilisée dans la méthode de krigeage plus loin) mais elle suffit ici puisque nous montrons assez vite qu'elle ne permet pas de résoudre notre problème.

Ainsi, nous définissons $V$ la matrice de Vandermonde associé au vecteur $(X_1,\ldots,X_N)$
$$
V = \begin{pmatrix}
1&X_1&\ldots&X_1^{N-1}\\
\vdots&\vdots&\vdots\\
1&X_N&\ldots&X_N^{N-1}
\end{pmatrix}.
$$
Le polynôme interpolateur de Lagrange associé aux points $X$ et $Y$ s'écrit alors
$$ P(X) = \sum_{i=0}^{N-1} \alpha_i X^i $$
où le vecteur $\alpha=(\alpha_0,\ldots,\alpha_{N-1})$ est solution de $V\alpha=Y$.

**Question**

> * Proposez une fonction `interp_Lagrange` qui prend en arguments 3 `ndarray` contenant les points d'interpolation $(X_1,\ldots,X_N)$, $(Y_1,\ldots,Y_N)$ et les points $(x_1,\ldots,x_n)$ où le polynôme $P$ est évalué et qui retourne un `ndarray` $y$ tel que $y_i = P(x_i)$ pour $1\leq i\leq n$.
> * Testez cette fonction en prenant pour points d'interpolation les points $X$ et $Y$ donnés en début de notebook.
> * Expliquez pourquoi l'interpolation par un polynôme ne permet pas d'avoir une bonne représentation des données.

```python
def interp_Lagrange(X,Y,x):
    n = len(X)
    V = np.ones((n,n))
    for i in np.arange(n):
        V[:,i] = X**i
    
    alpha = np.linalg.solve(V,Y)
    
    y = np.zeros(x.shape)
    for a in alpha[::-1]:
        y = y*x + a
    
    return y
```

```python
a, b, c, d = -5, 5, -1, 2

y = interp_Lagrange(X,Y,x)

fig = plt.figure(figsize=(6, 6))
ax = fig.add_subplot(1, 1, 1)
ax.scatter(X, Y, s=50, color='navy')
ax.plot(x, y)
ax.set_xlim(a, b)
ax.set_ylim(c, d)
ax.set_title('Nuage de points')
```

#### Sur les bords, on remarque que les polynomes tendent vers plus l'infini, notamment pour les polynomes de degrès elevés, en effet sur les bords, comme le polynôme est elevé et plus on s'éloigne de notre approximation. 


#### En effet, on constate que l'on approche bien la courbe au centre de l'intervalle , par contre , aux extrémités se produisent des oscillations parasites appelées effets de bord

# Méthode de krigeage



**Question**

> * Définissez les différentes fonctions $g$ (puissance, sphérique, exponentielle, gaussienne) proposées dans le cours.
> * Ajoutez l'effet pépite en pondérant comme une combinaison linéaire convexe les fonctions.
> * Affichez les sur un même graphique.

```python
def g_puissance(h,m,a, pepite = 0):
    if a == np.inf:
        out = 1-h
    else:
        out = (1- (h/a)**m) * (h<a)
    if pepite !=0:
        out[h>0] *= 1-pepite
    return out
```

```python
def g_spherique(h,a, pepite = 0):
    if a == np.inf:
        out = 1- 1.5*(h) +0.5*((h**3))
    else:
        out = (1- 1.5*(h/a) +0.5*((h**3)/(a**3)) ) * (h<a)
    if pepite !=0:
        out[h>0] *= 1-pepite
    return out
```

```python
def g_exponentiel(h,a, pepite = 0):

    out = np.exp(-h/a)
    if pepite !=0:
        out[h>0] *= 1-pepite
        
    return out
```

```python
def g_gaussien(h,a, pepite = 0):
    out = np.exp(-(h/a)**2)
    if pepite !=0:
        out[h>0] *= 1-pepite
        
    return out
```

```python
palier = 1
h = np.linspace(0,2*palier, 1000)
fig = plt.figure(figsize=(6,6))
ax = fig.add_subplot(1,1,1)

for pepite in [0.25]:
    ax.plot(h, g_puissance(h,1,palier,pepite=pepite),
    label =f'Puissance (m = {1} & pepite = {pepite})', linewidth=2)
    
    ax.plot(h, g_spherique(h,palier,pepite=pepite),
    label =f'Spherique (pepite = {pepite})', linewidth=2)
    
    ax.plot(h, g_exponentiel(h,palier,pepite=pepite),
    label =f'exponentiel (pepite = {pepite})', linewidth=2)
    
    ax.plot(h, g_gaussien(h,palier,pepite=pepite),
    label =f'Gaussien (pepite = {pepite})', linewidth=2)
    

ax.legend()
ax.set_title("Superposition des modèles")
plt.show()
```

#### On remarque qu'il y a un décalage de 0.25 en (0,0) qui est du à la pépite( c'est un décalage de 0.25 en y de la taille de la pépite)

**Question**

> * Proposez une fonction `krigeage` sur le même modèle que la fonction `interp_Lagrange` en lui passant également en paramètre le modèle choisi.
> * Affichez les différentes interpolations obtenues pour différents modèles avec ou sans pépite. Vous testerez différentes valeurs pour le palier et vous proposerez un ou plusieurs modèles qui marchent correctement. 

```python
def krigeage(X,Y,x, modele):
    
    M = modele['M']
    g = modele['g']
    
    N = X.size
    
    if Y.size != N:
        print("erreur")
        return
    
    V = np.zeros((N+M,N+M))

    for k in range(M):
        V[:N, N+k] = X**k
        V[N+k, :N] = X**k
    V[:N,:N] = g(abs(X[:, np.newaxis] - X[ np.newaxis, :]))

    sndmb = np.zeros((N+M,))
    sndmb[:N] = Y
    
    sol = np.linalg.solve(V,sndmb)
    alpha, beta = sol[:N], sol[N:]
    
    #evaluation du plonome en x P(x) = sum
    y = np.zeros(x.shape)
    for a in beta[::-1]:
        y = y*x + a
    for i, alphai in enumerate(alpha):
        y += alphai*g(abs(x-X[i]))
    
    return y 
```

```python
def plot_courbes1(m1,m2,t1,t2): # pour voir avec et sans pepite pour un meme modele
    y1 = krigeage(X,Y,x,m1)
    y2 = krigeage(X,Y,x,m2)
    fig, (ax1, ax2) = plt.subplots(1,2, figsize=(20,7))
    fig.figsize=(8, 6)
    
    ax1.set_ylim(-1,1.5)
    ax1.set_xlim(-6,6)
    ax1.scatter(X, Y, s=50, color='navy')
    ax1.plot(x, y1, color='orange')
    ax1.set_title(t1)

    ax2.set_ylim(-1,1.5)
    ax2.set_xlim(-6,6)
    ax2.scatter(X, Y, s=50, color='navy')
    ax2.plot(x, y2, color='orange')
    ax2.set_title(t2)
    
```

```python
def plot_courbes(m,t,param,ax,first): # pour voir avec pepite pour les 4 modeles ou sans pour les 4 modeles
    value,text = param
    compteur =0
    for i in range(2):
        for j in range(2):
            y = krigeage(X,Y,x,m[compteur])
            ax[i][j].set_ylim(-1,1.5)
            ax[i][j].set_xlim(-6,6)
            if first:
                ax[i][j].scatter(X, Y, s=50, color='navy')
            ax[i][j].plot(x, y, label =f'{text}={value}')
            ax[i][j].set_title(t[compteur])
            ax[i][j].legend()
            compteur+=1
```
