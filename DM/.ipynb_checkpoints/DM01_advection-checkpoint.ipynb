# Comparaison de schémas pour l'advection

Nous considérons le problème posé sur un domaine périodique
$$
\left\lbrace\begin{aligned}
&\partial_t u + a \partial_x u = 0, && t>0, 0\leq x\leq 1,\\
&u(0, x) = u^0(x), && 0\leq x\leq 1,\\
&u(t, 0) = u(t, 1), && t\geq 0.
\end{aligned}\right.
$$

L'objectif est de comparer les simulations numériques obtenues par différents schémas. Pour les applications numériques, nous prendrons $a=1$.

```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
%config InlineBackend.figure_format = 'retina'
```

## Le maillage

Le maillage en espace est constitué de $N$ points équi-répartis entre 0 et 1. Nous noterons
$$ x_j = h(j+1/2), \qquad 0\leq j< N,$$
avec $h=1/N$.

**Question**

> * Proposez une fonction `mesh` qui prend en argument un entier $N$ et qui retourne un `double` $h$ pour le pas du maillage et un `ndarray` de taille $N$ contenant les valeurs $x_j$ pour $0\leq j<N$.
> * Représentez le maillage obtenu avec $N=10$ en utilisant la commande `scatter`.

```python
def mesh(N):
    """ 
    h = 1/N #pas du maillage
    m = np.arange(0, 1, 1/N)
    m+=0.5*h"""
   
    """h = 1/N
    m = np.linspace(0,N-1,N) # car n exclu
    #print(m)
    m += 0.5
    m*=h
    #print(m)"""
    
    
    """construit le maillage pour les différences finies"""
    m, h = np.linspace(0, 1, N, endpoint=False, retstep=True)
    m += .5*h
    
    return h,m
```

```python
N = 10
fig = plt.figure(figsize=(6, 6))
ax = fig.add_subplot(1, 1, 1)
h, m = mesh(N)
ax.scatter(m, np.zeros(m.shape), s=50, c='orange')
ax.set_title(f'Maillage avec {N} points');
```

## La donnée initiale

Pour fonction initiale $u^0$, nous souhaitons pouvoir tester le schéma avec différentes régularités (une fonction discontinue, continue, dérivable, $\ldots$).

**Question**

> * Proposez une fonction `init` qui prend en argument un `ndarray` $x$ et un `integer` $k$ et qui retourne un `ndarray` $y$ tel que $y_i=\phi^k(x_i)$ où $\phi^k$ est définie par
$$
\phi^k (x) = \boldsymbol{1}_{[x_l,x_r]}(x) P^{k+1}(x),
\quad
P(x) = \frac{4}{(x_r-x_l)^2} (x-x_l)(x_r-x),
$$
avec $x_l=1/4$ et $x_r=1/2$.
> * Tracez dans un même graphique les différentes données initiales pour $k\in\lbrace -1, 0, 1, 2\rbrace$.

```python
def init(x,k):
    
    y = np.zeros(len(x))
    xl, xr = 0.25, 0.5
    compt=0
    for xi in x:
        
        p = (4/(xr-xl)**2) *(xi-xl)*(xr-xi)
        #p = 4/(xr-xl)**2 * (xi-xl)*(xr-xi)
        if(xi>=xl and xi<=xr):
            y[compt] = p**(k+1)
        else:
            y[compt] = 0
        compt+=1
    return y
```

```python
N = 1000
fig = plt.figure(figsize=(6, 6))
ax = fig.add_subplot(1, 1, 1)
h, m = mesh(N)
liste_k = [-1, 0, 1, 2]
n = len(liste_k)
for i, k in enumerate(liste_k):
    ax.plot(m, init(m, k), color=(i/n, 0, 1-i/n), linewidth=2, label=f'$k={k}$')
ax.set_title(f'Données initiales')
ax.legend();
```

## Les schémas

Nous allons à présent implémenter plusieurs schémas. Le schéma de Lax-Fridrichs (déjà vu en cours et en TD) défini par
$$
v_j^{n+1} = \frac{1}{2} (1+\beta) v_{j-1}^n + \frac{1}{2} (1-\beta) v_{j+1}^n,
$$
le schéma décentré à gauche défini par
$$
v_j^{n+1} = \beta v_{j-1}^n + (1-\beta) v_j^n,
$$
et le schéma de Lax-Wendroff
$$
v_j^{n+1} = \frac{1}{2} \beta(1+\beta) v_{j-1}^n + (1-\beta^2) v_j^n + \frac{1}{2} \beta(-1+\beta) v_{j+1}^n,
$$
avec $\beta=ak/h$, où $k$ est le pas de temps et $h$ le pas d'espace.

**Question**

> * Proposez des fonctions `lax_friedrichs`, `decentre` et `lax_wendroff` qui prennent en argument un `ndarray` $v$, un `double` $\beta$ et qui retourne un `ndarray` $w$ calculé après un pas de temps du schéma correspondant. Vous n'oublierez pas de traiter les termes de bord grâce à la condition de périodicité.
> * Testez les schémas proposé en traçant la solution approchée après un tour, 2 tours, 3 tours. Vous choisirez pour cela convenablement le paramètre $k$ en fonction de $h$ et de $a$. 
> * Faites varier la valeur de $ak/h$ afin de mesurer l'impact de la condition CFL.
> * Commentez les figures obtenues.

```python
v = np.array([0,1,2,3,4,5])
vp, vm = np.empty(v.shape), np.empty(v.shape)
vp[:-1] = v[1:]
vm[1:] = v[:-1]
# conditions aux limites
vp[-1] = v[0]
vm[0] = v[-1]

# j+1
print(vp) 

#  j-1 
print(vm) 
```

```python
def lax_friedrichs(v,beta):
    vp, vm = np.zeros(v.shape), np.zeros(v.shape) # vp est decale à droite soit j+1
    #vm est décalé à gauche j-1
    
    """
    for i in range(len(v)-1):
        vp[i] = v[i+1]
        
    vp[-1] = v[0]
     
    
    for i in range(1,len(v)):
        vm[i] = v[i-1]
    
    vm[0] = v[-1]"""

    vp[:-1] = v[1:]
    vm[1:] = v[:-1]
    # conditions aux limites
    vp[-1] = v[0]
    vm[0] = v[-1]
    alpham, alphap = .5*(1+beta), .5*(1-beta)
    
    return alpham*vm + alphap*vp
```

```python
def decentre(v,beta):
    vm = np.zeros(v.shape)
    # vp est decale à droite soit j+1
    #vm est décalé à gauche j-1
    """
    for i in range(len(v)-1):
        vp[i] = v[i+1]
        
    vp[-1] = v[0]
    """
    vm[1:] = v[:-1]
    vm[0] = v[-1]
            
    return beta*vm + (1-beta )* v.copy()
```

```python
def lax_wendroff(v,beta):
    vp, vm = np.zeros(v.shape), np.zeros(v.shape)
    # vp est decale à droite soit j+1
    #vm est décalé à gauche j-1
    
    """
    for i in range(len(v)-1):
        vp[i] = v[i+1]
        
    vp[-1] = v[0]
     
    
    for i in range(1,len(v)):
        vm[i] = v[i-1]
    
    vm[0] = v[-1]"""

    vp[:-1] = v[1:] 
    vm[1:] = v[:-1]
    
    vp[-1] = v[0]
    vm[0] = v[-1]
    
    alpham, alphav,alphap  =  .5*beta*(1+beta), (1-beta**2) ,.5*beta*(-1+beta)
    
    w = vp*alphap + vm*alpham + v.copy()*alphav
    
    return w
```

```python
a = 1  # vitesse
kk = 2  # paramètre de régularité
N = 200  # nombre de points pour le maillage
beta = 0.75  # paramètre pour la condition CFL
nb_tours = 3  # nombre de tours

h, m = mesh(N)  # maillage
k = beta/abs(a)*h  # pas de temps
nb_pas = int(1/(a*k))  # nombre de pas de temps pour faire un tour

fig, ax= plt.subplots(1, 3,figsize=(20,7))
fig.suptitle('Comparaison du nombre de tour des différents schémas', fontsize=16)
t = 0.
v = init(m, kk)

#ax.plot(m, v, color=(0, 0, 1), linewidth=2, label=f'$t={t:1.0f}$')

f_to_eval= ['lax_friedrichs','decentre','lax_wendroff'] # Tableau des schémas

for i,fte in enumerate(f_to_eval): # boucle par rapport auxschemas
    t = 0.
    v = init(m, kk)
    ax[i].plot(m, v, color=(0, 0, 1), linewidth=2, label=f'$t={t:1.0f}$ ') # pour t = 0 en bleu ( sol exact )
    for nt in range(1, nb_tours+1): # boucle pour faire un nombre de tours particulier
        for _ in range(nb_pas): # boucle pour faire un nombre de pas particulier pour 
            v = eval(fte+"(v, beta)")
            t += k
        ax[i].plot(m, v, color=(nt/nb_tours, 0, 1-nt/nb_tours), linewidth=2, label=f'$t={t:1.0f}$ ')

    ax[i].set_title(f'Solution par ' + fte)
    ax[i].legend();
```

### Plus le nombre de tour augmente, et plus l'erreur se propage, en effet, les courbes t = 1 à t = 3 les courbes sont moins porches/collées de la solution exacte t = 0. Sauf pour Lx wendroff pour laquelle les solution semblent s'approcher de la solution exacte, cependant contrairement aux deux autres schéma, le min et max n'est pas respecté. Decentre semble être plus proche de la solution exacte.

```python

a = 1  # vitesse
kk = 2  # paramètre de régularité
N = 200  # nombre de points pour le maillage
beta = 0.75  # paramètre pour la condition CFL
nb_tours = 4  # nombre de tours

h, m = mesh(N)  # maillage
k = beta/abs(a)*h  # pas de temps
nb_pas = int(1/(a*k))  # nombre de pas de temps pour faire un tour

fig, ax= plt.subplots(1, 3,figsize=(20,7))
fig.suptitle('Comparaison du nombre de Beta des différents schémas', fontsize=16)

t = 0.
v = init(m, kk)

#ax.plot(m, v, color=(0, 0, 1), linewidth=2, label=f'$t={t:1.0f}$')



f_to_eval= ['lax_friedrichs','decentre','lax_wendroff'] #Les schémas
for i,fte in enumerate(f_to_eval): # on boucle sur les schémas
    ax[i].plot(m, v, color=(0, 0, 0), linewidth=2, label= "exacte ")
    for beta in [0.1,0.3,0.5,0.7,0.9]: # on prends différentes valeurs de beta ici
        t = 0.
        k = beta/abs(a)*h
        nb_pas = int(1/(a*k))  
        v = init(m, kk) # on resete à 0
        for _ in range(nb_pas):#nombre de pas pour faire un tour uniquement ici
            v = eval(fte+"(v, beta)")
            t += k
        ax[i].plot(m, v, color=(nt/nb_tours, 0, 1-beta/0.9), linewidth=2, label= "beta " + str(beta))
    
    ax[i].set_title(f'Solution par ' + f_to_eval[i])
    ax[i].legend();
```

### Plus beta est grand, et plus la bosse est basse aplati et étalée, et donc on s'éloigne de la solution exacte. Plus beta est proche de 1 et plus les solutions sont précises d'après le schéma. 

### Donc on remarque que Lax wendorf n'est pas borné entre le min et le max  (ne respecte pas CFL )de la solution exacte, en effet on passe au dessus et en dessous, quelque soit Beta (sauf pour la solution exacte). Cependant Lax wendorf reste plus fidèle à la solution. Et decentre a l'air mieux que lax friedrichs car plus fidèle à la solution exact à chaque fois, de plus les deux gardent les propriétés de min et max de la solutionn exacte (condition CFL).

Nous allons à présent visualiser la convergence du schéma. Nous choisirons $\beta=0.75$ et des pas d'espace $h$ de la forme $h_n=2^{-n}$ pour $n$ allant de 4 à 11. Pour $h=h_n$, nous noterons $e_n$ la norme infinie de l'erreur (différence entre la solution exacte et la solution approchée) au bout d'un tour.

**Question**

> * Tracez le logarithme de $e_n$ en fonction de $h_n$ pour chacun des schémas.
> * Que pouvez-vous en conclure ?

```python
a = 1  # vitesse
kk = 2  # paramètre de régularité
beta = 0.75

liste_h = [[],[],[]]
liste_e = [[],[],[]]
f_to_eval= ['lax_friedrichs','decentre','lax_wendroff'] ##'lax_friedrichs','decentre','lax_wendroff'

for i,fte in enumerate(f_to_eval):
    
    for n in range(4,11):

        N = 3*2**n  # nombre de points pour le maillage
        beta = 0.75  # paramètre pour la condition CFL
        nb_tours = 4  # nombre de tours

        h, m = mesh(N)  # maillage
        k = beta/abs(a)*h  # pas de temps
        nb_pas = int(1/(a*k))  # nombre de pas de temps pour faire un tour

        t = 0.
        v = init(m,kk)
        
        for nt in range(1, nb_tours+1): # boucle pour faire un nombre de tours particulier
            for _ in range(nb_pas): # on fait un tour ici
                v = eval(fte+"(v, beta)")
                t += k
        
        en = np.linalg.norm(v -init(m,kk), np.inf)
        liste_h[i].append(np.log(h))
        liste_e[i].append(np.log(en))



fig = plt.figure(figsize=(6, 6))
ax = fig.add_subplot(1, 1, 1)


for i,fte in enumerate(f_to_eval):
    ax.scatter(liste_h[i], liste_e[i],label=f'erreur courbe: '+ fte)
    p = ((liste_e[i][-1])- (liste_e[i][-2])) / (liste_h[i][-1]-liste_h[i][-2])
    print(p) # laissé pour voir les pentes
    ax.plot(liste_h[i], p*(liste_h[i]-liste_h[i][-1])+liste_e[i][-1], label=  fte + f' $pente={p:1.0f}$ ')

ax.set_title(f'Logarithme de l erreur ' +'Erreurs' )
ax.legend();
```

### Plus il y a de points de le maillage et moins il n'y a d'erreur dans les schémas
### Plus il y a de tours et moins les pentes sont importantes et donc moins erreurs convergent
### Plus beta est proche proche de 1,( plus le nombre de tour augmente) et plus l'erreur se propage. Le schéma de wendroff converge plus rapidement car sa pente est plus élevée. On peut voir que la pente de firedrichs est legerement mopins importante que decentre.
### Lax wendroff semble avoir une meilleure approche, sa pente est en effet plus fidèle mais cependant elle borne mal. ( ne respecte pas le principe du maximum et du minimum ). Tandis que les deux autres, respectent les bornes min et max (CFL), mais leurs erreurs sont moins convergentes.
