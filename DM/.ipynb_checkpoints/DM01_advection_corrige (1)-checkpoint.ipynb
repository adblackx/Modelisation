# Comparaison de schémas pour l'advection

Nous considérons le problème posé sur un domaine périodique
$$
\left\lbrace\begin{aligned}
&\partial_t u + a \partial_x u = 0, && t>0, 0\leq x\leq 1,\\
&u(0, x) = u^0(x), && 0\leq x\leq 1,\\
&u(t, 0) = u(t, 1), && t\geq 0.
\end{aligned}\right.
$$

L'objectif est de comparer les simulations numériques obtenues par différents schémas. Pour les applications numériques, nous prendrons $a=1$.

```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
%config InlineBackend.figure_format = 'retina'
```

## Le maillage

Le maillage en espace est constitué de $N$ points équi-répartis entre 0 et 1. Nous noterons
$$ x_j = h(j+1/2), \qquad 0\leq j< N,$$
avec $h=1/N$.

**Question**

> * Proposez une fonction `mesh` qui prend en argument un entier $N$ et qui retourne un `double` $h$ pour le pas du maillage et un `ndarray` de taille $N$ contenant les valeurs $x_j$ pour $0\leq j<N$.
> * Représentez le maillage obtenu avec $N=10$ en utilisant la commande `scatter`.

```python
def mesh(N):
    """construit le maillage pour les différences finies"""
    m, h = np.linspace(0, 1, N, endpoint=False, retstep=True)
    m += .5*h
    return h, m
```

```python
N = 10
fig = plt.figure(figsize=(6, 6))
ax = fig.add_subplot(1, 1, 1)
h, m = mesh(N)
ax.scatter(m, np.zeros(m.shape), s=50, c='orange')
ax.set_title(f'Maillage avec {N} points');
```

## La donnée initiale

Pour fonction initiale $u^0$, nous souhaitons pouvoir tester le schéma avec différentes régularités (une fonction discontinue, continue, dérivable, $\ldots$).

**Question**

> * Proposez une fonction `init` qui prend en argument un `ndarray` $x$ et un `integer` $k$ et qui retourne un `ndarray` $y$ tel que $y_i=\phi^k(x_i)$ où $\phi^k$ est définie par
$$
\phi^k (x) = \boldsymbol{1}_{[x_l,x_r]}(x) P^{k+1}(x),
\quad
P(x) = \frac{4}{(x_r-x_l)^2} (x-x_l)(x_r-x),
$$
avec $x_l=1/4$ et $x_r=1/2$.
> * Tracez dans un même graphique les différentes données initiales pour $k\in\lbrace -1, 0, 1, 2\rbrace$.

```python
def init(x, k):
    xl, xr = 0.25, 0.5
    y = 1. * (x>xl) * (x<xr)
    p = 4/(xr-xl)**2 * (x-xl)*(xr-x)
    for _ in range(k+1):
        y *= p
    return y
```

```python
N = 1000
fig = plt.figure(figsize=(6, 6))
ax = fig.add_subplot(1, 1, 1)
h, m = mesh(N)
liste_k = [-1, 0, 1, 2]
n = len(liste_k)
for i, k in enumerate(liste_k):
    ax.plot(m, init(m, k), color=(i/n, 0, 1-i/n), linewidth=2, label=f'$k={k}$')
ax.set_title(f'Données initiales')
ax.legend()
plt.savefig("fonction_bosse.jpg")
```

## Les schémas

Nous allons à présent implémenter plusieurs schémas. Le schéma de Lax-Fridrichs (déjà vu en cours et en TD) défini par
$$
v_j^{n+1} = \frac{1}{2} (1+\beta) v_{j-1}^n + \frac{1}{2} (1-\beta) v_{j+1}^n,
$$
le schéma décentré à gauche défini par
$$
v_j^{n+1} = \beta v_{j-1}^n + (1-\beta) v_j^n,
$$
et le schéma de Lax-Wendroff
$$
v_j^{n+1} = \frac{1}{2} \beta(1+\beta) v_{j-1}^n + (1-\beta^2) v_j^n + \frac{1}{2} \beta(-1+\beta) v_{j+1}^n,
$$
avec $\beta=ak/h$, où $k$ est le pas de temps et $h$ le pas d'espace.

**Question**

> * Proposez des fonctions `lax_friedrichs`, `decentre` et `lax_wendroff` qui prennent en argument un `ndarray` $v$, un `double` $\beta$ et qui retourne un `ndarray` $w$ calculé après un pas de temps du schéma correspondant. Vous n'oublierez pas de traiter les termes de bord grâce à la condition de périodicité.
> * Testez les schémas proposé en traçant la solution approchée après un tour, 2 tours, 3 tours. Vous choisirez pour cela convenablement le paramètre $k$ en fonction de $h$ et de $a$. 
> * Faites varier la valeur de $ak/h$ afin de mesurer l'impact de la condition CFL.
> * Commentez les figures obtenues.

```python
def lax_friedrichs(v, beta):
    """un pas de temps du schéma de Lax-Friedrichs"""
    vr, vl = np.empty(v.shape), np.empty(v.shape)
    vr[:-1] = v[1:]
    vl[1:] = v[:-1]
    # conditions aux limites
    vr[-1] = v[0]
    vl[0] = v[-1]
    alphal, alphar = .5*(1+beta), .5*(1-beta)
    return alphal*vl + alphar*vr

def decentre(v, beta):
    """un pas de temps du schéma décentré à gauche"""
    vr, vl = np.empty(v.shape), np.empty(v.shape)
    vr[:-1] = v[1:]
    vl[1:] = v[:-1]
    # conditions aux limites
    vr[-1] = v[0]
    vl[0] = v[-1]
    alphal, alpham = beta, (1-beta)
    return alphal*vl + alpham*v

def lax_wendroff(v, beta):
    """un pas de temps du schéma de Lax-Wendroff"""
    vr, vl = np.empty(v.shape), np.empty(v.shape)
    vr[:-1] = v[1:]
    vl[1:] = v[:-1]
    # conditions aux limites
    vr[-1] = v[0]
    vl[0] = v[-1]
    alphal, alpham, alphar = .5*beta*(1+beta), 1-beta**2, .5*beta*(-1+beta)
    return alphal*vl + alpham*v + alphar*vr
```

```python
a = 1  # vitesse
kk = 3  # paramètre de régularité
N = 300  # nombre de points pour le maillage
beta = 0.5  # paramètre pour la condition CFL
nb_tours = 3  # nombre de tours

h, m = mesh(N)  # maillage
k = beta/abs(a)*h  # pas de temps
nb_pas = int(1/(a*k))  # nombre de pas de temps pour faire un tour
if 1/(a*k) - nb_pas != 0.0:
    print("Problème : le nombre de pas pour faire un tour n'est pas entier")

liste_scheme = [lax_friedrichs, decentre, lax_wendroff]
liste_name = ['Lax-Friedrichs', 'Decentré', 'Lax-Wendroff']
ns = len(liste_scheme)

fig = plt.figure(figsize=(5*ns, 5))

for ks, scheme in enumerate(liste_scheme):
    ax = fig.add_subplot(1, ns, ks+1)
    t = 0.
    v = init(m, kk)
    ax.plot(m, v, color=(0, 0, 1), linewidth=2, label=f'$t={t:1.0f}$')
    for nt in range(1, nb_tours+1):
        for _ in range(nb_pas):
            v = scheme(v, beta)
            t += k
        ax.plot(m, v, color=(nt/nb_tours, 0, 1-nt/nb_tours), linewidth=2, label=f'$t={t:1.0f}$')
    ax.set_title(f'Solution par {liste_name[ks]}')
    ax.legend()

plt.savefig(f"schema_{kk}.pdf")
```

**Commentaires sur les résultats obtenus**

Le comportement du *schéma décentré* est très similaire à celui du schéma de Lax-Friedrichs. La solution numérique semble vérifier le **principe du maximum** lorsque $\beta$ est compris entre $0$ et $1$. Il y a de la **diffusion numérique** (la courbe est applatie, le pic perd de l'amplitude et la base s'élargie au fur et à mesure des tours). Mais le point positif est qu'il semble que cette diffusion numérique soit moindre.

*Remarque :* il est possible de démontrer que le principe du maximum est vérifié exactement comme pour le schéma de Lax-Friedrichs. Le terme $v_j^{n+1}$ s'écrit comme une combinaison linéaire convexe lorsque $\beta\in[0, 1]$.

Le comportement du *schéma de Lax-Wendroff* est très différent. Tout d'abord, le **principe du maximum** n'est pas vérifié (on peut également trouver des contre-exemples assez simples mais visuellement, on observe que la solution approchée peut prendre des valeurs négatives) et nous observons des oscillations (qui n'ont rien à voir avec la solution réelle du problème) derrière la bosse et pas devant. A ce niveau, il n'est pas demandé d'explications du pourquoi ! Egalement, la **diffusion numérique** semble beaucoup plus faible puisque nous n'observons pas vraiment de baisse d'amplitude ou d'élargissement de la bosse au fur et à mesure des tours.

Il n'est pas très raisonnable de décider ainsi quel est le "meilleur" schéma. Chacun a des propriétés positives et négatives. Tout va dépendre des applications qui nous intéressent pour choisir tel ou tel schéma.

Nous allons à présent visualiser la convergence du schéma. Nous choisirons $\beta=0.75$ et des pas d'espace $h$ de la forme $h_n=2^{-n}$ pour $n$ allant de 4 à 11. Pour $h=h_n$, nous noterons $e_n$ la norme infinie de l'erreur (différence entre la solution exacte et la solution approchée) au bout d'un tour.

**Question**

> * Tracez le logarithme de $e_n$ en fonction de $h_n$ pour chacun des schémas.
> * Que pouvez-vous en conclure ?

```python
a = 1  # vitesse
kk = 2  # paramètre de régularité
beta = 0.75  # paramètre pour la condition CFL
nb_tours = 1  # nombre de tours

liste_scheme = [lax_friedrichs, decentre, lax_wendroff]
liste_name = ['Lax-Friedrichs', 'Decentré', 'Lax-Wendroff']
ns = len(liste_scheme)

fig = plt.figure(figsize=(5*ns, 5))

for ks, scheme in enumerate(liste_scheme):
    ax = fig.add_subplot(1, ns, ks+1)

    liste_n = list(range(4, 12))
    liste_en = []
    liste_hn = []
    for n in liste_n:
        N = 3*2**n
        h, m = mesh(N)  # maillage
        k = beta/abs(a)*h  # pas de temps
        nb_pas = int(1/(a*k))  # nombre de pas de temps pour faire un tour
        v = init(m, kk)
        for _ in range(nb_pas):
            v = scheme(v, beta)
        #en = np.sqrt(h)*np.linalg.norm(v - init(m, kk))
        en = np.linalg.norm(v - init(m, kk), np.inf)
        liste_en.append(en)
        liste_hn.append(h)

    liste_hn = np.asarray(liste_hn)
    ax.scatter(liste_hn, liste_en, s=50, c='red')
    ax.plot(liste_hn, liste_en[-1]/liste_hn[-1]*liste_hn, linewidth=2, c='navy')
    ax.plot(liste_hn, liste_en[-1]/liste_hn[-1]**2*liste_hn**2, linewidth=2, c='orange')
    ax.set_xscale('log', base=2)
    ax.set_yscale('log', base=2)
    ax.set_title(f"Evolution de l'erreur pour {liste_name[ks]}")
    
plt.savefig("convergence.pdf")
```

**Commentaires sur les résultats obtenus**

Nous rappelons que nous essayons de comparer l'erreur à une puissance de $h$ (le pas d'espace qui est proportionnel au pas de temps $k$ car $\beta$ est choisi constant dans ces simulations). Nous essayons d'écrire que 
$$ \vert e(h) \vert \leq C h^p,$$
où $p$ est une constante que l'on cherche à déterminer.
Nous traçons donc $\log(\vert e(h)\vert)$ en fonction de $h$ et l'on cherche une droite de pente $p$ telle que les points soient placés sous la droite.

Pour le *schéma de Lax-Friedrichs* et le *schéma décentré*, la droite a une pente de $1$. Donc l'erreur se comporte comme $h$, c'est-à-dire qu'elle tend vers 0 comme $h$ lorsque $h$ tend vers 0.
Pour le *schéma de Lax-Wendroff*, en revanche, la pente semble être 2 (numériquement on est plutôt aux alentours de $1.9$). Ainsi l'erreur tend comme $h^2$ lorsque $h$ tend vers 0, ce qui est beaucoup plus rapide. La solution proposée par le schéma de Lax-Wendroff converge beaucoup plus vite vers la solution continue : pour une précision donnée, ce schéma nécessite un maillage beaucoup moins fin, donc moins de calcul.
